## 1. Import Data

```python
# importing data
path = r"C:\Users\bhimc\Documents\Instacart Basket Analysis (Career Foundry Rekha)"
```

```python
# importinf data set orders_cleaned
df_ords = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_cleaned.csv'), index_col = False)
```

```python
# imporing data set orders_prior
df_ords_prior = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'orders_products_prior.csv'), index_col = False)
```

## 3. Merge Dataframe df_ords and df_ords_prior

```python
#Create a new dataframe merged
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
# Check output of Merged dataframe
df_merged_large.head()
```

```python
# After merge Check the frequency
df_merged_large['_merge'].value_counts()
```

```python
# Export merged dataframe to pickle 'pkl'
df_merged_large.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```
